```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Load the entire dataset with only required columns to save memory and process faster
cols = ['uf', 'condicao_metereologica', 'sentido_via', 'causa_acidente', 'tipo_acidente', 'municipio', 'mortos', 'classificacao_acidente']
df=pd.read_csv('https://docs.google.com/uc?export=download&id=16dnWpVpcanPOizgwzaTBcjNbK7VIqZ6r',sep=';',encoding='latin1')

# Filter for fatal accidents
# We can use df['mortos'] > 0 or df['classificacao_acidente'] == 'Com VÃ­timas Fatais'
df_fatais = df[df['mortos'] > 0]

```

## Quais sÃ£o as 10 UFs com mais acidentes fatais?

```python
# --- Grafico 1: UFs mais ocorrem acidentes fatais ---
plt.figure()
uf_fatais = df_fatais['uf'].value_counts().head(10)
sns.barplot(x=uf_fatais.values, y=uf_fatais.index, palette='Reds_r')
plt.title('Top 10 UFs com Mais Acidentes Fatais (2025)', fontsize=14, fontweight='bold')
plt.xlabel('NÃºmero de Acidentes Fatais')
plt.ylabel('UF')
plt.tight_layout()
plt.savefig('grafico1_uf_fatais.png')
plt.show()
```

Minas Gerais (MG) lidera com folga o ranking de acidentes com mortes, seguido por ParanÃ¡ (PR) e Bahia (BA).

Fato curioso: Apesar de SÃ£o Paulo ter uma das maiores frotas, o rigor e as caracterÃ­sticas das rodovias paulistas (muitas sob concessÃ£o) a deixam na 10Âª posiÃ§Ã£o em fatalidades registradas pela PRF.

## Qual condiÃ§Ã£o meteorolÃ³gica ocorrem mais acidentes?

```python
# --- Grafico 2: Qual condiÃ§Ã£o meteorolÃ³gica ocorrem mais acidentes (Geral) ---
plt.figure()
cond_met = df['condicao_metereologica'].value_counts().head(8)
sns.barplot(x=cond_met.values, y=cond_met.index, palette='Blues_r')
plt.title('Acidentes por CondiÃ§Ã£o MeteorolÃ³gica (Geral)', fontsize=14, fontweight='bold')
plt.xlabel('NÃºmero de Acidentes')
plt.ylabel('CondiÃ§Ã£o MeteorolÃ³gica')
plt.tight_layout()
plt.savefig('grafico2_condicao_meteorologica.png')
plt.show()
```

A esmagadora maioria dos acidentes acontece com o tempo em condiÃ§Ãµes ideais (CÃ©u Claro). Isso indica que a imprudÃªncia ou a distraÃ§Ã£o humana pesam muito mais do que fatores climÃ¡ticos adversos, jÃ¡ que o tempo bom traz uma falsa sensaÃ§Ã£o de seguranÃ§a que induz ao excesso de velocidade.

## Qual sentido da via ocorrem mais acidentes?

```python
plt.figure()
sentido = df['sentido_via'].value_counts().head(6)
sns.barplot(x=sentido.values, y=sentido.index, palette='viridis')
plt.title('Acidentes por Sentido da Via (Geral)', fontsize=14, fontweight='bold')
plt.xlabel('NÃºmero de Acidentes')
plt.ylabel('Sentido da Via')
plt.tight_layout()
plt.savefig('grafico3_sentido_via.png')
plt.show()
```

O sentido Crescente da via registra um volume ligeiramente maior de acidentes em comparaÃ§Ã£o ao sentido Decrescente.

## Qual a causa de acidentes mais comum?

```python
plt.figure()
causa = df['causa_acidente'].value_counts().head(10)
sns.barplot(x=causa.values, y=causa.index, palette='flare')
plt.title('Top 10 Causas Mais Comuns de Acidentes', fontsize=14, fontweight='bold')
plt.xlabel('NÃºmero de Acidentes')
plt.ylabel('Causa do Acidente')
plt.tight_layout()
plt.savefig('grafico4_causa_acidente.png')
plt.show()
```

A principal causa registrada Ã© a AusÃªncia de reaÃ§Ã£o do condutor (distraÃ§Ã£o, uso de celular), seguida bem de perto por ReaÃ§Ã£o tardia ou ineficiente do condutor. Juntas, as falhas de atenÃ§Ã£o dominam as ocorrÃªncias.

## Qual o tipo de acidentes que mais ocorre?

```python
plt.figure()
tipo = df['tipo_acidente'].value_counts().head(10)
sns.barplot(x=tipo.values, y=tipo.index, palette='crest')
plt.title('Top 10 Tipos de Acidentes Mais Frequentes', fontsize=14, fontweight='bold')
plt.xlabel('NÃºmero de Acidentes')
plt.ylabel('Tipo de Acidente')
plt.tight_layout()
plt.savefig('grafico5_tipo_acidente.png')
plt.show()
```

A ColisÃ£o traseira Ã© o tipo de acidente disparado mais comum (geralmente ligada Ã  falta de distÃ¢ncia de seguranÃ§a e reaÃ§Ãµes tardias), seguida por SaÃ­da de leito carroÃ§Ã¡vel (quando o veÃ­culo sai da pista) e ColisÃ£o transversal.

## Quais sÃ£o 10 municÃ­pios que mais tem acidentes fatais?

```python
plt.figure()
mun_fatais = df_fatais['municipio'].value_counts().head(10)
sns.barplot(x=mun_fatais.values, y=mun_fatais.index, palette='magma')
plt.title('Top 10 MunicÃ­pios com Mais Acidentes Fatais', fontsize=14, fontweight='bold')
plt.xlabel('NÃºmero de Acidentes Fatais')
plt.ylabel('MunicÃ­pio')
plt.tight_layout()
plt.savefig('grafico6_municipios_fatais.png')
plt.show()
```
